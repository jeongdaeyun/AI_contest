```python
import torch

DEVICE = torch.device('mps:0' if torch.backends.mps.is_available() else 'cuda:0' if torch.cuda.is_available() else 'cpu')
print(f"your device: {DEVICE}")

model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
model.to(DEVICE)

imgs = ['./data/temp/temp1.jpeg']

results = model(imgs)

# results.print()
results.show()

# results.xyxy[0] # tensor
results.pandas().xyxy[0] # pandas
```
